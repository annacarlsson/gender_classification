```python
# This files creates a test set from the Adience dataset
from pathlib import Path
from collections import defaultdict
from math import inf
import numpy as np
import os
import csv
```

```python
dataset = Path('/home/deeplearn/Desktop/Adience/')
output = Path('Sorted')
file = 'fold_4_data.txt' # We use one of the folds

counts = defaultdict(int)

def process(row):
    file_name = row[1]
    gender = row[5]
    if counts[file_name] > 1:
        return
    
    if gender.isdigit():
        gender = row[4]
        
    if gender == 'f':
        dest = output / 'Female'
    elif gender == 'm':
        dest = output / 'Male'
    else:
        print(gender)
        return
    
    src = dataset / 'aligned' / Path(row[0]) 
    os.makedirs(dest, exist_ok=True)
    file = list(src.glob('*' + file_name))[0]
    try:
        os.symlink(src / file, dest / file_name)
    except:
        os.remove(dest / file_name)
        os.symlink(src / file, dest / file_name)

def check(row):
    file_name = row[1]
    counts[file_name] += 1
    
with open(dataset / file) as fin:
    next(fin)
    rows = (line.split() for line in fin)
    for row in rows:
        check(row)
        
with open(dataset / file) as fin:
    next(fin)
    rows = (line.split() for line in fin)
    for row in rows:
        process(row)
#with open(dataset / file) as csvfile:
#    reader = csv.DictReader(csvfile, delimiter='\t')
#    for row in reader:
#        check(row)
        
#with open(dataset / file) as csvfile:        
#    reader = csv.DictReader(csvfile, delimiter='\t')
#    for row in reader:
#        process(row)
```
